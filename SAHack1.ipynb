```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('hacktrain.csv')
test=pd.read_csv('hacktest.csv')
```

```python

x_train=train.iloc[:,3:31]
y_train=train['class']
test=test.iloc[:,2:30]
```

```python
x_train=x_train.interpolate(axis=1,limit_direction='both')
```

```python
x_train.isnull().sum()
test.isnull().sum()
```

```python
x_train.describe()
```

```python
for col in x_train.columns:
    quantile1,quantile3=np.percentile(x_train[col],[25,75])
    iqr=quantile3-quantile1
    lower_bound_val = quantile1 -(1.5 * iqr) 
    upper_bound_val = quantile3 +(1.5 * iqr) 
    x_train[col] = x_train[col].where((x_train[col] >= lower_bound_val) & (x_train[col] <= upper_bound_val), np.nan)
    test[col] = test[col].where((test[col] >= lower_bound_val) & (test[col] <= upper_bound_val), np.nan)
```

```python
x_train.isnull().sum()
test.isnull().sum()
```

```python
x_train=x_train.interpolate(axis=1,limit_direction='both')
test=test.interpolate(axis=1,limit_direction='both')
```

```python
x_train.isnull().sum()
```

```python
for col in x_train.columns:
    quantile1,quantile3=np.percentile(x_train[col],[25,75])
    iqr=quantile3-quantile1
    lower_bound_val = quantile1 -(1.5 * iqr) 
    upper_bound_val = quantile3 +(1.5 * iqr) 
    x_train[col] = np.clip(x_train[col], lower_bound_val, upper_bound_val)
    test[col] = np.clip(test[col], lower_bound_val, upper_bound_val)
    
```

```python
x_train.describe()
```

```python
test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
test=scaler.fit_transform(test)
```

```python
y_train.map({'water':0,'forest':1,'impervious':2,'farm':3,'grass':4,'orchard':5})
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(x_train,y_train)
```

```python
prediction=classifier_regressor.predict(test)
```

```python
test_df=pd.read_csv('hacktest.csv')
ids=test_df['ID']
```

```python
submission = pd.DataFrame({
    "ID": ids,
    "class": prediction
})
```

```python
submission.to_csv("Hack1.csv", index=False)
```
